```python
# Task#3: Function that checksif number is Even or Odd
def Number_Checker():
    number = int(input("Enter a number: "))
    if number % 2 == 0:
        print("Number is Even")
    else:
        print("Number is Odd")

for i in range(10):
    Number_Checker()


```
